```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
data
```

```python
Austria = data[data.country == 'Austria']
Austria
```

```python
Bahrain = data[data.country == 'Bahrain']
Bahrain
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Austria.year, Austria.population / 10**6)
plt.plot(Bahrain.year, Bahrain.population / 10**6)
plt.legend(['Austria', 'Bahrain'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Austria.year, Austria.population / Austria.population.iloc[0] * 100)
plt.plot(Bahrain.year, Bahrain.population / Bahrain.population.iloc[0] * 100)
plt.legend(['Austria', 'Bahrain'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```
